```python
# Step 2: Import the necessary modules
from transformers import pipeline
from PIL import Image
import requests
```

```python
# Step 3: Create an image classification pipeline
image_classifier = pipeline(task="image-classification")
```

```python
# Step 4: Define the image you want to classify
# You can use a URL of an image or upload an image to Colab and provide the file path
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/b/bb/Kittyply_edit1.jpg/1280px-Kittyply_edit1.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)
```

```python
# Step 5: Perform image classification
results = image_classifier(image)
```

```python
# Step 6: Display the classification results
for result in results:
    print(f"Label: {result['label']}, Score: {result['score']:.2f}")
```
