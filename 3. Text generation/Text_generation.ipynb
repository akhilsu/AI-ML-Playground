```python
# Step 2: Import the necessary modules
from transformers import pipeline
```

```python
# Step 3: Create a text generation pipeline
text_generator = pipeline(task="text-generation", model="gpt2")
```

```python
# Step 4: Define the prompt for text generation
prompt = "Once upon a time in a land far, far away"
```

```python
# Step 5: Generate text based on the prompt
generated_texts = text_generator(prompt, max_length=50, num_return_sequences=1)
```

```python
# Step 6: Display the generated text
for idx, generated_text in enumerate(generated_texts):
    print(f"Generated Text {idx + 1}: {generated_text['generated_text']}\n")
```
