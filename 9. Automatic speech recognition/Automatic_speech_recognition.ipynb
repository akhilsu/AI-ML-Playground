```python
# Step 2: Import the necessary modules
from transformers import pipeline
```

```python
# Step 3: Create an automatic speech recognition pipeline
asr_pipeline = pipeline(task="automatic-speech-recognition", model="facebook/wav2vec2-base-960h")
```

```python
# Step 4: Define the path to the local audio file
audio_path = "audio.mp3"
```

```python
# Step 5: Perform speech recognition
transcription = asr_pipeline(audio_path)
```

```python
# Step 6: Display the transcription result
print("Transcription:")
print(transcription['text'])
```
