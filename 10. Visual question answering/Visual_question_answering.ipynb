```python
# Step 2: Import the necessary modules
from PIL import Image
from transformers import pipeline
```

```python
# Step 3: Create a Visual Question Answering pipeline
vqa_pipeline = pipeline("visual-question-answering")
```

```python
# Step 4: Load the image and define the question
image = Image.open("elephant.jpeg")  # Ensure this file is uploaded to your Colab environment
question = "Is there an elephant?"
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
# Step 5: Perform VQA
result = vqa_pipeline(image, question, top_k=1)
```

```python
# Step 6: Display the result
print(result)
```
